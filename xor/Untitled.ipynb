```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y = np.array([1000, 20000, 4000, 50000])
t = np.array([100, 4000, 200000, 40000])
```

```python
error = np.absolute(Y-t)
error
```

```python
pd.DataFrame(error).plot()
plt.show()
```

```python
logerror = np.absolute(np.log(Y) - np.log(t))
logerror
```

```python
pd.DataFrame(logerror).plot()
plt.show()
```
